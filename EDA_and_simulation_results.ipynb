```python
# extract items per counter
import pandas as pd
import ast

def analyze_counters_and_items(row):
    col_str = row['col']
    
    # Handle NaN or missing data - updated to return float('nan') for the ratio
    if pd.isna(col_str):
        return pd.Series([0, 0, float('nan')])
    
    # Step 1: Parse the 'col' string into a Python object
    if isinstance(col_str, str):
        # Convert lowercase true/false to Python True/False
        py_format = col_str.replace('true', 'True').replace('false', 'False')
        try:
            col_data = ast.literal_eval(py_format)
        except Exception:
            return pd.Series([0, 0, float('nan')])
    else:
        col_data = col_str

    # Step 2: Extract counts from the col tuple structure: (cid, q_len, items, is_open)
    open_counters_count = 0 
    items_count = 0
    
    for cid, q_len, items, is_open in col_data:
        if is_open:
            open_counters_count += 1
            items_count += items
            
    # Step 3: Calculate the ratio safely, returning NaN if no counters are open
    if open_counters_count > 0:
        ratio = items_count / open_counters_count
    else:
        ratio = float('nan')

    return pd.Series([open_counters_count, items_count, ratio])
```

```python
# compute the confidence interval
import pandas as pd
import numpy as np
import scipy.stats as st

# 1. Load and convert the data ONCE (outside the loop)
customer_df_all = log_converter.apply(log_Customers, variant=log_converter.Variants.TO_DATA_FRAME)
counter_df_all = log_converter.apply(log_counter, variant=log_converter.Variants.TO_DATA_FRAME)

counter_df_all = counter_df_all.rename(columns={'case:concept:name': 'cid'})
customer_df_all['time:timestamp'] = pd.to_datetime(customer_df_all['time:timestamp'])
counter_df_all['time:timestamp'] = pd.to_datetime(counter_df_all['time:timestamp'])

customer_df_all = customer_df_all.sort_values('time:timestamp')
counter_df_all = counter_df_all.sort_values('time:timestamp')

# Extract target events once
target_events_all = customer_df_all[customer_df_all["concept:name"].isin(["Go to Checkout"])].copy()

# 2. Run the loop over the selected hours
for h in [14, 15, 16, 17, 18, 19, 20, 21]:
    print(f"--- Processing Hour {h}:00 ---")
    
    # Strictly filter BOTH dataframes to the current hour before merging
    counter_df = counter_df_all[counter_df_all['time:timestamp'].dt.hour == h]
    target_events = target_events_all[target_events_all['time:timestamp'].dt.hour == h]
    
    # Handle edge case: if either dataframe is empty for this hour, skip it
    if target_events.empty or counter_df.empty:
        print(f"No data available for hour {h}.\n")
        continue

    # Merge matching logs chronologically
    events_with_queue_state = pd.merge_asof(
        target_events,
        counter_df[['time:timestamp', 'col']],  
        on='time:timestamp',
        direction='forward'                    
    )
    events_with_queue_state = events_with_queue_state.rename(columns={'col_y': 'col'})
    
    # Apply your metric calculations
    events_with_queue_state[['open_counters', 'items_count', 'ratio']] = events_with_queue_state.apply(
        analyze_counters_and_items, axis=1
    )

    # Clean data array for bootstrapping
    data_array = events_with_queue_state['ratio'].dropna().values

    # Check if we have enough data points to bootstrap
    if len(data_array) < 2:
        print(f"Not enough data points in hour {h} to calculate Bootstrap CI.\n")
        continue

    # 3. Calculate Bootstrap Statistics
    sample_median = np.median(data_array)

    bootstrap_result = st.bootstrap(
        data=(data_array,), 
        statistic=np.median, 
        confidence_level=0.95,
        n_resamples=9999,  # 9,999 is standard and significantly faster than 99,999
        method='percentile'
    )

    low = bootstrap_result.confidence_interval.low
    high = bootstrap_result.confidence_interval.high

    # --- PRINT RESULTS ---
    print(f"Sample Median: {sample_median:.2f}")
    print(f"95% Bootstrap CI for Median: [{low:.2f}, {high:.2f}]\n")
```

```python
# extract if the queue length of a counter is higher than 3
def analyze_counters_and_queues(row):
    col_str = row['col']
    
    # Handle NaN or missing data
    if pd.isna(col_str):
        return pd.Series([0, 0, 0.0])
    
    # Step 1: Parse the 'col' string into a Python object
    if isinstance(col_str, str):
        # Convert lowercase true/false to Python True/False
        py_format = col_str.replace('true', 'True').replace('false', 'False')
        try:
            col_data = ast.literal_eval(py_format)
        except Exception:
            return pd.Series([0, 0, 0.0])
    else:
        col_data = col_str

    # Step 2: Extract counts from the col tuple structure: (cid, q_len, items, is_open)
    open_counters_count = 0 
    long_queues_count = 0
    
    for cid, q_len, items, is_open in col_data:
        if is_open:
            open_counters_count += 1
            if q_len >= 3:
                long_queues_count += 1
            
    # Step 3: Calculate the ratio safely to avoid ZeroDivisionError
    ratio = long_queues_count / open_counters_count if open_counters_count > 0 else 0.0

    return pd.Series([open_counters_count, long_queues_count, ratio])
```
